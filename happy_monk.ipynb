```python
import pandas as pd
import numpy as np
from sklearn import datasets 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tabulate import tabulate as tb
import matplotlib as plt
from matplotlib import pyplot
```

```python
bank_note = pd.read_csv('BankNote_Authentication.csv')
iris = datasets.load_iris()
breast_cancer = datasets.load_breast_cancer()
mnist = datasets.load_digits()
```

```python
#Target Class Analysis
print(set(bank_note['class']))
print(set(iris.target))
print(set(breast_cancer.target))
print(set(mnist.target))
```

```python
bank_note_df = pd.DataFrame(bank_note)
iris_df = pd.DataFrame(iris.data)
iris_df['target'] = iris.target
breast_cancer_df = pd.DataFrame(breast_cancer.data)
breast_cancer_df['target'] = breast_cancer.target
mnist_df = pd.DataFrame(mnist.data)
mnist_df['target'] = mnist.target
```

```python
print(bank_note_df.shape)
print(iris_df.shape)
print(breast_cancer_df.shape)
print(mnist_df.shape)
```

```python
dataset = [bank_note_df, iris_df, breast_cancer_df, mnist_df] 
dataset_name = ["Bank Note", "Iris", "Breast Cancer", "MNIST"]
```

```python
for i,j in zip(dataset, dataset_name):
  print(str(j))
  print(tb(i.head(10), headers = 'keys', tablefmt = 'fancy_grid'))
```

```python
for i,j in zip(dataset, dataset_name):
  print(f'Model for {j}:')
  X, y = i.values[ : , : -1], i.values[ : , -1]
  X = X.astype('float32')
  y = LabelEncoder().fit_transform(y)
  features_count = set(y) 
  if len(features_count) == 2:
    activation_function  = "sigmoid"
    d = 1 
    v_loss ='binary_crossentropy'
  else:
    activation_function = "softmax"
    d = len(features_count)
    v_loss = 'sparse_categorical_crossentropy'
  print('Target Values :- ',d)
  print(f'Activation function is {activation_function}')
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
  n_features = X.shape[1]
  model = Sequential()
  model.add(Dense(n_features*10, activation='relu', kernel_initializer='he_normal', input_shape=(n_features,)))
  model.add(Dense(d, activation=activation_function))
  print(model.summary())
  model.compile(optimizer='adam', loss = v_loss, metrics=['accuracy'])
  history = model.fit(X_train, y_train, epochs=100, batch_size=50, verbose=2, validation_data=(X_test,y_test))
  yhat_classes = model.predict_classes(X_test, verbose=1)
  # yhat_classes = yhat_classes[:, 0]
  # print('test predict',yhat_classes)
  # print('test true',y_test)
  if activation_function  == "sigmoid":
    yhat_classes = yhat_classes[:, 0]
  else:
    yhat_classes = yhat_classes
  # print(yhat_classes)
  train_loss, train_results = model.evaluate(X_train,y_train, verbose=2)
  test_loss, test_results = model.evaluate(X_test, y_test, verbose=2)
  f1 = f1_score(y_test, yhat_classes, average='micro')
  print('Train loss : ', (train_loss*100))
  print('Train accuracy : ', (train_results*100))
  print('Test loss : ', (test_loss*100))
  print('Test/Validation accuracy : ', (test_results*100))
  print('F1 score: %f' % (f1*100))
  pyplot.title('Loss')
  pyplot.plot(history.history['loss'], label='train')
  pyplot.plot(history.history['val_loss'], label='test')
  pyplot.legend()
  pyplot.show() 
  pyplot.title('Model Accuracy')
  pyplot.plot(history.history['accuracy'], label='train')
  pyplot.plot(history.history['val_accuracy'], label='test')
  pyplot.ylabel('accuracy')
  pyplot.xlabel('epoch')
  pyplot.legend(['train','test'], loc = 'upper right')
  pyplot.show() 
  pyplot.title('Model Loss')
  pyplot.plot(history.history['loss'], label='train')
  pyplot.plot(history.history['val_loss'], label='test')
  pyplot.ylabel('loss')
  pyplot.xlabel('epoch')
  pyplot.legend(['train','test'], loc = 'upper right')
  pyplot.show() 
```
